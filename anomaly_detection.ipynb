```python
import anomaly_detection_functions as adf
```

```python
contratos_df = adf.read_partitioned_csv("etl_data/contratos_infraesructura_df.csv")
contratos_df.head()
```

```python
directorio = "clustering_data"
nombre_archivo = "contratistas_clustered_df.csv"
delimitador = "," 

contratistas_clustered_df = adf.leer_csv(directorio, nombre_archivo, delimitador)
contratistas_clustered_df = contratistas_clustered_df[["ID_CONTRATISTA","cluster"]]

contratistas_clustered_df = contratistas_clustered_df.drop_duplicates(subset=["ID_CONTRATISTA"])


contratistas_clustered_df.head()
```

```python
directorio = "clustering_data"
nombre_archivo = "indice_transparencia.csv"
delimitador = "," 

indice_transparencia_df = adf.leer_csv(directorio, nombre_archivo, delimitador)
indice_transparencia_df = indice_transparencia_df[["cluster","indice_de_transparencia"]]
indice_transparencia_df.head()
```

```python
contratistas_indice = contratistas_clustered_df.merge(indice_transparencia_df, on="cluster", how="left")
contratistas_indice.head()
```

```python
contratos_indice = contratos_df.merge(contratistas_indice, on="ID_CONTRATISTA", how="left")
contratos_indice.head()
```

```python
contratos_indice.columns

contratos_indice.index = contratos_indice['REFERENCIA_CONTRATO']
contratos_indice = contratos_indice.drop(columns=['REFERENCIA_CONTRATO'])
contratos_indice.head()


numerical_columns = ["VALOR_TOTAL_CONTRATO"]
categorical_columns = ["TIPO_CONTRATO", "NOMBRE_GRUPO", "NOMBRE_FAMILIA", "NOMBRE_CLASE"]

preprocessed_contratos = adf.preprocess_data(contratos_indice, numerical_columns, categorical_columns)
preprocessed_contratos["cluster"] = contratos_indice["cluster"]

preprocessed_contratos.head()

```

```python
features = preprocessed_contratos.columns

anomalous_data = adf.detect_anomalies(preprocessed_contratos, 'cluster', features)

anomalous_data["original_index"] = anomalous_data.index
anomalous_data = anomalous_data[["anomaly_score"]]

anomalous_data.head()
```

```python
contratos_indice.head()
```

```python
merged_data = contratos_indice.merge(anomalous_data, left_on='REFERENCIA_CONTRATO', right_on='original_index', how='left')
merged_data.head()
```

```python
merged_data["indice_de_transparencia_del_contrato"] = merged_data["indice_de_transparencia"]*(1-merged_data["anomaly_score"])
merged_data.head()
```

```python
merged_data.to_csv("anomaly_data/contratos_indice_transparencia.csv")
```
